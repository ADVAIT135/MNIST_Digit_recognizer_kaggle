## Data Preparation and preprocessing

```python
pd.set_option('display.max_rows', 999)
round(train.drop('label', axis=1).mean(), 2).sort_values(ascending = False)
```

#### Average ie mean varies between 0 to 140. So, we Scale them

```python
# splitting into X and y
X = train.drop("label", axis = 1)
y = train['label']
```

```python
# scaling the features
X_scaled = scale(X)

# train test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, train_size=0.2,test_size = 0.8, random_state = 101)
```

```python
print('X_train shape:',X_train.shape)
print('y_train shape:',y_train.shape)
print('X_test shape:',X_test.shape)
print('y_test shape:',y_test.shape)
```

## Training with SVC - Linear Model

```python
model_linear = SVC(kernel='linear')
model_linear.fit(X_train, y_train)
```

```python
y_pred = model_linear.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn import metrics
print("accuracy:", metrics.accuracy_score(y_true=y_test, y_pred=y_pred))
```

## So, we got an accuracy of 91.3125% using SVC - Linear Model

## Viewing the Confusion Matrix

```python
print(metrics.confusion_matrix(y_true=y_test, y_pred=y_pred))
```

```python
print(metrics.confusion_matrix(y_true=y_test, y_pred=y_pred_poly_kern))
```

## (B) - Training with **rbf kernel**

```python
rbf_ker_mod = SVC(kernel='rbf')
rbf_ker_mod.fit(X_train, y_train)
y_pred_rbf_ker_mod = rbf_ker_mod.predict(X_test)
```

### Viewing the Accuracy, Confusion Matrix and Classification Report

```python
print(classification_report(y_test,y_pred_rbf_ker_mod))
```

```python
print("accuracy:", metrics.accuracy_score(y_true=y_test, y_pred=y_pred_rbf_ker_mod))
```

```python
print(metrics.confusion_matrix(y_true=y_test, y_pred=y_pred_rbf_ker_mod))
```

## So, we got a Accuracy of 93.961% using SVC Non Linear - rbf kernel

### The Accuracy jumped from 91.312% to 93.961% 
### So, we will be using SVC Non Linear - rbf kernel for training, also the precision values are high 
### i.e. above 90%(0.90) in almost all cases in rbf kernel

## **GRID SEARCH : HyperParameter Tuning**

#### Creating a Kfold object with 5 splits

```python
folds = KFold(n_splits = 5, shuffle = True, random_state = 101)
```

```python
# specifying range of hyperparameters
# Setting the parameters by cross-validation
hyper_params = [ {'gamma': [0.01, 0.001,0.0001],
                     'C': [1, 10, 100]}]
```

```python
final_model = SVC(kernel="rbf")
```

#### Setting up GridSearchCV()

```python
model_cv = GridSearchCV(estimator = final_model, 
                        param_grid = hyper_params, 
                        scoring= 'accuracy', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True,n_jobs = -1)  
```

```python
model_cv.fit(X_train, y_train)
```

```python
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
# converting C to numeric type for plotting on x-axis
cv_results['param_C'] = cv_results['param_C'].astype('int')

# # plotting
plt.figure(figsize=(20,7))

# subplot 1/3
plt.subplot(131)
gamma_01 = cv_results[cv_results['param_gamma']==0.01]

plt.plot(gamma_01["param_C"], gamma_01["mean_test_score"])
plt.plot(gamma_01["param_C"], gamma_01["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Gamma=0.01")
plt.ylim([0.50, 1.1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')

plt.subplot(132)
gamma_001 = cv_results[cv_results['param_gamma']==0.001]

plt.plot(gamma_001["param_C"], gamma_001["mean_test_score"])
plt.plot(gamma_001["param_C"], gamma_001["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Gamma=0.001")
plt.ylim([0.50, 1.1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')

plt.subplot(133)
gamma_0001 = cv_results[cv_results['param_gamma']==0.0001]

plt.plot(gamma_0001["param_C"], gamma_0001["mean_test_score"])
plt.plot(gamma_0001["param_C"], gamma_0001["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Gamma=0.0001")
plt.ylim([0.50, 1.1])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')
```

```python
# printing the optimal accuracy score and hyperparameters
best_score = model_cv.best_score_
best_hyperparams = model_cv.best_params_

print("The best test score is {0} corresponding to hyperparameters {1}".format(best_score, best_hyperparams))
```

### Building and Evaluating Final model

```python
rbf_final_model = SVC(C=10, gamma=0.001, kernel="rbf")

rbf_final_model.fit(X_train, y_train)
y_pred = rbf_final_model.predict(X_test)

# metrics
print("accuracy", metrics.accuracy_score(y_test, y_pred), "\n")
print(metrics.confusion_matrix(y_test, y_pred), "\n")
print(classification_report(y_test,y_pred)) 
```
